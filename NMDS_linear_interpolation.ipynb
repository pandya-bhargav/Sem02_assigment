## 3. Estimate the natural logarithm of 2 using linear interpolation. First, perform
## the computation by interpolating between In 1 =0 and In 6 = 1.791759.
## Then, repeat the procedure, but use a smaller interval from In 1 to In 4
## (1.386294). Note that the true value of In 2 is 0.6931472.
## Try for some different intervals and show errors for all intervals.
## Write your observation for the relationship between error and intervals.

```python
print("first Enter x : ")

x = float(input("x = "))

n = int(input("how many iteration you want =  "))

x_ = []
f_x = []


for i in range(n):
    ele = float(input("Enter x  = "))
    f_ele = float(input("Enter f_ele = "))
    x_.append( ele)
    f_x.append(f_ele)

print("x" , x_)
print("f_x" ,f_x)
```

```python
def linear_interpolation(x , x0 , f_x0 , y , f_y ):
  ans = f_x0 + (x-x0)*((f_x0-f_y)/(x0-y))
  return ans 

pred = []

for i in range(1,n):
    print("for ele ")
    print((x_[0],f_x[0]) , " and " , (x_[i], f_x[i]))
    

    pred_val = linear_interpolation(x , x_[0] , f_x[0] , x_[i] , f_x[i] )
    
    pred.append(pred_val)

    def per_error(pred_val , y=math.log(x)):
      return ((y-pred_val)/y)*100

    error = per_error(pred_val , math.log(x))
    print("predicted value = {}".format( pred_val ))
    print("Actual Value = {}" .format(math.log(x)) )
    print("Error =  {} %".format(round(error,4)))
    
    print("__________________________________________________________________________________________")
```

As we decrease the length of the interval, then the error is also reduced, and our interpolation will be more accurate.
