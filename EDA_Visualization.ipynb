### Why use graphs to present data?

Because they…

* are quick and direct,
*highlight the most important facts,
*facilitate understanding of the data,
*can convince readers,
*can be easily remembered.

### A good graph:

* accurately shows the facts,
*grabs the reader’s attention,
*complements or demonstrates arguments presented in the text,
*has a title and labels,
*is simple and uncluttered,
*shows data without altering the message of the data,
*clearly shows any trends or differences in the data,
*is visually accurate (i.e. if one chart value is 15 and another 30, then 30 should appear to be twice the size of 15).

### different types of graphs:

* bar charts (comparison of items and relationships, time series, frequency distribution)
*pictographs
*pie charts (description of components)
*line charts (time series and frequency distribution)
*scatterplots (analysis of relationships)
*histograms




## Plots for categorical data 
are used to represent data that can be grouped into categories or classes. Some commonly used plots for categorical data are:

>`Bar charts:` Bar charts are used to represent categorical data in the form of vertical or horizontal bars. Each bar represents a category and the height or length of the bar represents the frequency or proportion of data in that category.

>`Pie charts:` Pie charts are circular graphs divided into slices, with each slice representing a category. The size of each slice is proportional to the frequency or proportion of data in that category.

>`Stacked bar charts:` Stacked bar charts are similar to regular bar charts, but instead of representing a single category, they represent multiple categories stacked on top of each other. This type of chart is useful for comparing the relative frequencies of different categories.

>`Pareto charts:` Pareto charts are a combination of `bar` and `line charts`, with the bars representing the frequency or proportion of data in each category and the line representing the cumulative frequency or proportion.

>`Heatmaps:` Heatmaps are used to represent categorical data in a matrix format, with each row and column representing a category and the color or shading of each cell representing the frequency or proportion of data in the corresponding category pair.

>`Mosaic plots:` Mosaic plots are a combination of `stacked bar charts` and `heatmaps`, with each row and column representing a category and the area of each rectangle representing the frequency or proportion of data in the corresponding category pair.

>`Dot plots:` Dot plots are used to represent categorical data in the form of dots or points. Each dot represents a single data point, and the dots are arranged according to their categories. This type of plot is useful for detecting outliers and identifying patterns in the data.

The choice of plot depends on the type of data and the research question being investigated.

## Plots for categorical data 
>`Histograms:` Histograms are a common way to visualize the distribution of a numeric variable. The x-axis represents the range of values for the variable, and the y-axis represents the frequency of values within each range.

>`Box plots:` Box plots display the distribution of a numeric variable, showing the median, quartiles, and outliers. They are useful for comparing distributions between groups or identifying unusual values.

>`Scatter plots:` Scatter plots display the relationship between `two numeric variables`. Each point represents the values for the two variables for a single observation. They can be used to identify trends, clusters, or outliers.

>`Line charts:` Line charts display the relationship between two variables over time. They are useful for identifying trends and patterns over time.

>`Heat maps:` Heat maps use color to represent the value of a numeric variable across two dimensions. They are useful for identifying patterns in large datasets.

>`Density plots:` Density plots display the distribution of a numeric variable as a smooth curve. They are similar to histograms, but provide a more continuous view of the distribution.

>`Violin plots:` Violin plots display the distribution of a numeric variable, similar to box plots, but provide a more detailed view of the shape of the distribution.

# BAR CHART
* either horizontal or vertical. 
* the greater the bar length or height, the greater their value. 
* recognize patterns or trends.
* usually present categorical (nominal) variables, discrete variables or continuous variables grouped in class intervals. 
* consist of an axis and a series of labelled horizontal or vertical bars. 
* bars depict frequencies, proportion of different values of a variable
* bars visualize changes over time
* compare multiple variables simultaneously
* numbers on the y-axis of a vertical bar chart or the x-axis of a horizontal bar chart are called the scale.


## Vertical bar charts
* useful to compare different categorical or discrete variables, such as age groups, classes, schools, etc., as long as there are not too many categories to compare. 
* very useful for time series data. 
*The space for labels on the x-axis is small, but ideal for years, minutes, hours or months. 

## Guidelines to build bar charts
* Make bars and columns wider than the space between them.
* Use a single font type on a chart. 
* Try to maintain a consistent font style from chart to chart in a single presentation or document.
* Order your shade pattern from darkest to lightest.
* Avoid using a combination of red and green in the same display.

# Scatter plot 
* used to present measurements of two or more related variables. 
* particularly useful when the values of the variables of the y-axis are thought to be dependent upon the values of the variable of the x-axis.
* data points are plotted but not joined. 
* indicates the type and strength of the relationship between two or more variables. 
  * a linear or non-linear relationship
    * When data points form a straight line on the graph, the relationship between the variables is linear
    * When data points don’t form a line or when they form a line that is not straight, the relatioship is nonlinear
  * a positive (direct) or negative (inverse) relationship,
    * If the points cluster around a line that runs from the lower left to upper right of the graph area, then the relationship between the two variables is said to be positive or direct
    * If the points cluster around a line that runs from the upper left to the lower right of the graph area, then the relationship is said to be negative or inverse
  *the concentration or spread of data points,
    * Data points can be close together or spread widely across the graph area
  *the presence of outliers
    * Outliers are data points that are far from the other points in the data set

# Box plot
* also caleed the `box and whisker plot`
* used to compare univariate data
* help visualize the five-number summary.
  * The box stretches from the lower hinge (Q1) to the upper hinge (Q3).
  * The median (Q2) is shown as a line across the box. 
  * The "H-spread/Hinge length" is `the difference between the hinges`. 
  * A "step" is defined as `1.5 times` the `H-spread`.
  * `Inner fences` are 1 step beyond the hinges. 
  * `Outer fences` are 2 steps beyond the hinges.
  * There are two adjacent values: 
    * the `largest value` below the upper inner fence 
    * the `smallest value` above the lower inner fence. 
  * a line is drawn from the 
    * `upper hinge` to the `upper adjacent value` and 
    * `lower hinge` to the `lower adjacent value`
  * Every score between the inner and outer fences is indicated by an "o" 
  * Every score beyond the outer fences is indicated by a "*".
* check a distribution's skewness
* check potential unusual observations (outliers) in the data set
  * Outliers are greater than Q3+(1.5 . IQR) or less than Q1-(1.5 . IQR).
* ideal for comparing distributions because the centre, spread and overall range are immediately apparent.
*useful to compare data from two or more groups by viewing box plots from the groups side by side.

## Variable width box plot
* illustrate the size of each group whose data is being plotted by making the `width` of the box proportional to the `size` of the group. 
* A popular convention is to make the `box width` proportional to the `square root of group size`.

## Notched box plot 
* apply a "notch" or narrowing of the box around the `median`. 
* Notches are useful in offering a rough guide of the significance of the difference of medians; 
  * if the notches of two boxes do not overlap, this will provide evidence of a statistically significant difference between the medians.
* The width of the notches is 
  * proportional to IQR of the sample 
  * inversely proportional to the square root of sample size

* The notch boundaries are at a distance of $\pm \frac{1.58~IQR}{\sqrt{n}}$ from the median

## Adjusted box plot
* describe skew distributions
* rely on the medcouple (MC) statistic of skewness
* lengths of the upper and lower whiskers on the box-plot are respectively 
  * $ 1.5~ IQR~e^{3 MC}, ~~1.5 ~IQR~ e^{-4 MC} \text{ if } MC > 0$,
  * $ 1.5~ IQR~e^{4 MC}, ~~1.5 ~IQR~ e^{-3 MC} \text{ if } MC < 0$.
* For a symmetric distribution, MC=0,$\implies$ adjusted box-plot reduces to box-plot.

## Violin plot
* A violin plot is a combination of a box plot and a kernel density plot. 
* it starts with a box plot and then adds a rotated kernel density plot to each side of the box plot.
* shows the full distribution of the data.
* For multimodal distributions, it shows the presence of different peaks, their position and relative amplitude.
* Individual outliers are not visible in a violin plot.

## Bean plot
*  A beanplot is an alternative to the boxplot for visual
comparison of univariate data between groups. 
* Construction:
  * individual observations are shown as small lines in a one-dimensional scatter plot. 
  * estimated density of the distributions is visible and the average is shown. 
* It is easy to compare
different groups of data in a beanplot and to see if a group contains enough observations
to make the group interesting from a statistical point of view. 
* Anomalies in the data, such
as bimodal distributions and duplicate measurements, are easily spotted in a beanplot.

#Dendrogram 
* is a network structure. 
* constitute of a `root node` that gives birth to several nodes connected by `edges or branches`. 
* last nodes of the hierarchy are called `leaves`. 


#Types
* Two type of dendrogram exist, resulting from 2 types of dataset:

  * A `hierarchic dataset` provides the links between nodes explicitely. 
  * A `clustering algorithm` can be visualized as a dendrogram.

# Variations
* It can be `horizontal` or `vertical`  
* It can be `linear` or `circular` (space efficient)
* display a heatmap at the bottom of the dendrogram

# Sketch
To draw a dendrogram, you first need to have 
* a numeric matrix. 
* Each line represents an entity. 
* Each column is a variable. 

The objective is to cluster the entities to show `who shares similarities with whom`. 
* The dendrogram will draw the similar entities closer to each other in the tree.

```python
# Libraries
from scipy.cluster.hierarchy import dendrogram, linkage
titanic.head()
```

```python
data1 = titanic[['survived','pclass','age','sibsp','parch','fare']]
data1.head()
#Z = linkage(titanic, 'ward')
```

```python
data1.info()
```

```python
data2=data1.dropna(axis=0).reset_index(drop=True)
data2.head()
```

```python
data2.info()
```

```python
merg = linkage(data2[0:10], method="ward")
print(merg)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
dendrogram(merg, leaf_rotation = 0)
ax.set_title('Hierarchical Clustering')
ax.set_xlabel('data points')
ax.set_ylabel('euclidean distance')
```
