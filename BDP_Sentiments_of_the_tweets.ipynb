# â€¢ Stream the tweets from handles of Yan Lee Cunn, Adani (20 tweets from each handles)

```python
class StreamClient(tweepy.StreamingClient):
    def __init__(self, bearer_token, keyword):
        self.limit = 20
        self.tweets = []
        self.keyword = keyword
        super(StreamClient, self).__init__(bearer_token)

    def on_tweet(self, tweet):
        self.tweets.append([self.keyword, tweet.text])
        print(f'{len(self.tweets)}. {tweet.text}')
        if len(self.tweets) >= self.limit:
            self.disconnect()
```

# â€¢ Stream the tweets containing keywords â€adaniâ€, â€unionbudgetâ€ and apply textblob to find the sentiments of the tweets.

```python
print('Tweets for keyword "Adani"')
adani_stream = StreamClient(bearer_token=bearer_token, keyword='adani')
adani_stream.delete_rules(adani_stream.get_rules().data)
adani_stream.add_rules(tweepy.StreamRule('adani'))

try:
    adani_stream.filter()
except Exception as e:
    print(e)

print()
print('Tweets for keyword "union budget"')
budget_stream = StreamClient(bearer_token=bearer_token, keyword='union budget')
budget_stream.delete_rules(budget_stream.get_rules().data)
budget_stream.add_rules(tweepy.StreamRule('union budget'))
try:
    budget_stream.filter()
except Exception as e:
    print(e)
```

```python
data = adani_stream.tweets
data.extend(budget_stream.tweets)
```

### Creating dataframe

```python
df = pd.DataFrame(data=data, columns=['Keyword', 'Tweet'])
df
```

### Calculating sentiment for each tweet

```python
def calculate_sentiment(text):
    try:
        polarity_score = TextBlob(text).sentiment.polarity
        if polarity_score <= -0.25:
          return 'Negative'
        elif polarity_score > -0.25 and polarity_score < 0.25:
          return 'Neutral'
        else:
          return 'Positive'
    except:
        return None

df['Sentiment'] = df['Tweet'].apply(calculate_sentiment)
df
```
