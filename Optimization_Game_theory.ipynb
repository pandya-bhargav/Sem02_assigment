```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = [[-4,2,5,-6,-6],
     [3,-9,7,4,8]]
```

```python
def saddle_point(A):
    row_min = [0]*len(A)
    for i in range(len(A)):
        row_min[i] = min(A[i])

    A = np.array(A)
    A = A.T
    col_max = [0]*len(A)
    for j in range(len(A)):
        col_max[j] = max(A[j])
    return max(row_min) == min(col_max)
```

```python
saddle_point(A)
```

```python
A = np.array(A)
A = A.T
x = np.linspace(0,1,10000)
for ele in A:
    y = ele[0]*x + ele[1]*(1-x)
    plt.plot(x,y,label = "{}*x + {}*(1-x)".format(ele[0] , ele[1]))

plt.legend(loc= 'best' ,fontsize='small' )
```

```python
new_A = [[2 , -6 ],[-9 , 4]]
```

```python
def find_coff(new_A):
    d = new_A[0][0]-new_A[1][0]-new_A[0][1]+new_A[1][1]
    dl = new_A[0][1]-new_A[1][1]
    l = -(dl/d)
    dk = -(new_A[1][0]-new_A[1][1])
    k = dk/d
    c = new_A[1][1]-d*k*l
    return d , l , k ,c
```

```python
d ,l , k , c = find_coff(new_A)
```
